```python
import pandas as pd

# Load the dataset
url = "https://huggingface.co/datasets/hmcgovern/original-language-bibles-greek/resolve/main/data/train-00000-of-00001.parquet"
df = pd.read_parquet(url)

# Filter for John 1:1 token rows (reference format: Jhn.1.1.01, Jhn.1.1.02, ...)
john_1_1 = df[df["reference"].astype(str).str.startswith("Jhn.1.1.")]

# Convert to dictionary format
verse_data = [
    {"greek": row["text"], "english": row["translation"]}
    for _, row in john_1_1.iterrows()
]

print(verse_data)
```
